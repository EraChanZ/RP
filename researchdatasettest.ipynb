```python
from researchdataset import VideoDataset
video_directory = "C:/Users/vladi/RP/Research/IR_videos"  # Replace with your directory
dataset = VideoDataset(video_directory, transform=None, frame_skip=240)
```

```python
# Get total size of frames in memory
import sys
total_frames = len(dataset.data)
if total_frames > 0:
    # Get size of first frame as sample
    frame_size = sys.getsizeof(dataset.data[0].tobytes())
    total_size_bytes = frame_size * total_frames
    
    # Convert to MB
    total_size_mb = total_size_bytes / (1024 * 1024)
    
    print(f"Number of frames: {total_frames}")
    print(f"Size per frame: {frame_size/1024:.2f} KB")
    print(f"Total memory usage: {total_size_mb:.2f} MB")
else:
    print("Dataset is empty")
```

```python
# Get a single frame and its metadata
frame, metadata = dataset[0]

# Import matplotlib for displaying the image
import matplotlib.pyplot as plt

# Display the frame
plt.figure(figsize=(10, 8))
plt.imshow(frame)
plt.title(f"Video: {metadata['video_name']}, Frame: {metadata['frame_num']}")
plt.axis('off')
plt.show()
```
